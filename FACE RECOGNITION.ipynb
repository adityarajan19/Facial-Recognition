# Implementing Face Recognition using Python

```python
import face_recognition

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import cv2
%matplotlib inline

```

```python
# STEP 1

# Loading the known images to make the machine learn it

image = cv2.imread('leo.jpg')
leo = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = cv2.imread('tom.jpg')
tom = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = cv2.imread('bill.jpg')
bill = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(leo)
```

```python
plt.imshow(tom)
```

```python
plt.imshow(bill)
```

```python
# STEP 2

# Getting the face encodings

leo_encoding = face_recognition.face_encodings(leo)[0]


tom_encoding = face_recognition.face_encodings(tom)[0]


bill_encoding = face_recognition.face_encodings(bill)[0]

# Creating a database of known face encodings

known_face_encodings = [
    leo_encoding,
    tom_encoding,
    bill_encoding
]
```

```python
# STEP 3 - Loading and Testing.

# Loading the unknown image 1 that we want to check

image = cv2.imread('leo2.jpg')
unknown_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(unknown_image)

# Getting face encodings of unknown image

unknown_face_encodings = face_recognition.face_encodings(unknown_image)
```

```python
# Testing if the unknown image matches any of the people we know

from scipy.spatial import distance

for unknown_face_encoding in unknown_face_encodings :

    results = []

for known_face_encoding in known_face_encodings :
    d = distance.euclidean(known_face_encoding, unknown_face_encoding)
    results.append(d)

threshhold = 0.6
results = np.array(results) <= threshhold

name = "unknown person"

if results[0]:
    name = "Lionel Messi from F.C Barcelona"
elif results[1]:
    name = "Tom Hanks"
elif results[2]:
    name = "Bill Gates"
    
print(f"Found {name} in the photo!")        
    
```
